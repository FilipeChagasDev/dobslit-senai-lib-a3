```python
from dobslit_senai_a3 import *
```

```python
class Exemplo1(GroverAlgorithm):
    def prepare(self):
        self.x1 = self.create_qubit('x1')
        self.x2 = self.create_qubit('x2')
        self.x3 = self.create_qubit('x3')
        self.t = self.create_qubit('t')
        
    def build_search_space(self):
        self.h(self.x1)
        self.h(self.x2)
        self.h(self.x3)
        # t = x1 and not x2 and x3
        self.logic_not(self.x2)
        self.logic_and([self.x1, self.x2, self.x3], self.t)
        self.logic_not(self.x2)
        
    def revert_search_space(self):
        self.logic_not(self.x2)
        self.logic_and([self.x1, self.x2, self.x3], self.t)
        self.logic_not(self.x2)
        self.h(self.x3)
        self.h(self.x2)
        self.h(self.x1)
        
ex1 = Exemplo1()
ex1.build_all(ex1.t, 1)
ex1.circ.draw()
```

```python
ex1.simulate()
```
